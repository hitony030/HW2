```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df['Formatted Date']=df['Formatted Date'].apply(lambda x:x[:11])
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
```

```python
df.info()
```

```python
df[df['Precip Type'].isna()].head()
```

```python
df[df['Precip Type']=='rain'].head()
```

```python
df[df['Precip Type']=='snow'].head()
```

```python
df['Precip Type'] = df['Precip Type'].fillna('warm')
```

```python
print('Number Of Missing Values: ',df.isna().sum().sum())
```

```python
df['year'] = df['Formatted Date'].dt.year
df['month'] = df['Formatted Date'].dt.month
df['day'] = df['Formatted Date'].dt.day
```

```python
df[['Formatted Date','year','month','day']].head()
```

```python
y = df['Temperature (C)']
x = df.drop(columns =['Temperature (C)','Formatted Date'])
```

```python
y.head()
```

```python
x.head()
```

```python
categorical_cols = x.select_dtypes(include='O')
numerical_cols = x.select_dtypes(include=[np.number])
```

```python
categorical_cols.dtypes
```

```python
numerical_cols.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
encoded_arr = encoder.fit_transform(categorical_cols)
```

```python
cols_name = categorical_cols.columns
cols_name
```

```python
encoded_arr
```

```python
encoded_df = pd.DataFrame(
    encoded_arr,
    columns=encoder.get_feature_names_out(cols_name)
)
```

```python
encoded_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_cols.loc[:,:] = scaler.fit_transform(numerical_cols)
```

```python
numerical_cols
```

```python
X = pd.merge(encoded_df,numerical_cols,left_index=True,right_index=True)
```

```python
X.shape
```

```python
# Define a Function For Data Visualization
def plot_data(actual,predicted,method,deg=1):
    F = pd.DataFrame({
        "Actual":actual,
        "Predicted":predicted
    })
    m,b=np.polyfit(F['Actual'],F['Predicted'],deg)
    plt.figure(figsize=(10,6))
    plt.scatter(F['Actual'],F['Predicted'],label='Data Points')
    plt.plot(F['Actual'],m*F['Actual']+b,color='red',label=f"Y={m:.2f}x+{b:.2f}")
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(method)
    plt.legend()
    plt.show()
```

```python
from sklearn.feature_selection import  SelectKBest,f_classif
kbest = SelectKBest(score_func=f_classif,k=20)
X = kbest.fit_transform(X,y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
lr = LinearRegression()
lr.fit(x_train,y_train)
train_pred = lr.predict(x_train)
test_pred = lr.predict(x_test)
print("========Train Result========")
print(f"Mean Absolute Error : {mean_absolute_error(y_train,train_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_train,train_pred)}")
print(f"R Square Score : {r2_score(y_train,train_pred)}")
print("========Test Result========")
print(f"Mean Absolute Error : {mean_absolute_error(y_test,test_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_test,test_pred)}")
print(f"R Square Score : {r2_score(y_test,test_pred)}")
```

```python
plot_data(y_test,test_pred,method='Linear Regression')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_fet = PolynomialFeatures(degree=2)
x_train_poly = poly_fet.fit_transform(x_train)
x_test_poly = poly_fet.fit_transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train_poly,y_train)
train_pred = lr.predict(x_train_poly)
test_pred = lr.predict(x_test_poly)
print("========Train Result========")
print(f"Mean Absolute Error : {mean_absolute_error(y_train,train_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_train,train_pred)}")
print(f"R Square Score : {r2_score(y_train,train_pred)}")
print("========Test Result========")
print(f"Mean Absolute Error : {mean_absolute_error(y_test,test_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_test,test_pred)}")
print(f"R Square Score : {r2_score(y_test,test_pred)}")
```

```python
plot_data(y_test,test_pred,method='Poly Linear Regression')
```
